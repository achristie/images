```python
import spgci as ci
from datetime import date
import pandas as pd
```

```python
mdd = ci.MarketData()
```

```python
symbols = ['RCUST04', 'RCCNT04', 'RCGMT04', 'RCUSL04', 'RCUSO04', 'RCUSG04', 'RCUSI04', 'RCUSM04', 'RCUSV04', 'RCUSD04', 'RCUSH04']
```

```python
mdd.get_assessments_by_symbol_current(symbol=symbols)
```

```python
ref = mdd.get_symbols(symbol=symbols)
ref.head()
```

```python
mdd.get_assessments_by_symbol_historical(symbol='RCUST04',)
```

```python
df = mdd.get_assessments_by_symbol_historical(assess_date_gte=date(2022,1,1), symbol=symbols).sort_values("assessDate", ascending=False).reset_index(drop=True)
df.head()
```

```python
df = df.merge(ref, on='symbol')
df.head()
```

```python
filt = df['assessDate'].isin(df.drop_duplicates("assessDate").reset_index(drop=True).iloc[:2]['assessDate'])
latest_df = df[filt]
latest_df.head()
```

```python
year_ago = df['assessDate'].max() - pd.DateOffset(years=1)
filt = df['assessDate'] >= year_ago
filt2 = df["assessDate"] == df[filt]['assessDate'].min()
year_ago_df = df[filt2]
year_ago_df.head()
```

```python
concat = pd.concat([latest_df, year_ago_df]).reset_index(drop=True)
concat.head()
```

```python
pt = concat.pivot_table(index=['description', 'symbol'], columns='assessDate', values='value').sort_values("assessDate", axis=1, ascending=False)
pt
```

```python
pt['Change WoW'] = pt.iloc[:, 0] - pt.iloc[:, 1]
```

```python
def sort_symbols(sym: str) -> int:
  d = { s: symbols.index(s) for s in sym}
  return d
```

```python
pt.sort_values("symbol", key=sort_symbols)
```
